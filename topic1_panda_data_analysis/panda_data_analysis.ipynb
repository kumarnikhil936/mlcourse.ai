```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

Dataset: Telecom Churn 

```python
df = pd.read_csv('telecom_churn.csv')
```

```python
print(df.shape)
```

```python
# To see statistics on numerical value
df.describe()
```

```python
df.info()
```

```python
# changing the datatype of a column
df['churn'] = df['churn'].astype('int64')
```

```python
# To see statitics on non numerical features
df.describe(include=['object', 'bool'])
# object = categorical type
```

```python
df['churn'].value_counts()
```

```python
df['churn'].value_counts(normalize=True)
```

```python
# Sorting
df.sort_values(by='total day charge', ascending=False).head()
```

```python
# Sorting by multiple columns
df.sort_values(by=['churn','total day charge'], ascending=False).head()
```

```python
# Indexing and retrieving data
df['churn'].mean()
```

```python
df[df['churn']==1].mean()
```

```python
# How much time (on average) do churned users spend on phone during daytime?

df[df['churn']==1]['total day minutes'].mean()
```

```python
# What is the maximum length of international calls among loyal users (Churn == 0) who do not have an international plan?

df[(df['churn']==0) & (df['international plan']=='no')]['total intl minutes'].max()
```

```python
# DataFrames can be indexed by column name (label) or row name (index) 
# or by the serial number of a row. The loc method is used for 
# indexing by name, while iloc() is used for indexing by number.


# "give us the values of the rows with index from 0 to 5 (inclusive) 
# and columns labeled from State to Area code (inclusive)"
df.loc[0:5, 'state':'area code']
```

```python
# "give us the values of the first five rows in the first three columns"
df.iloc[0:5, 0:3]
```

```python
# To apply functions to each column, use apply():
df.apply(np.max)
```

```python
# if we need to select all states starting with W, we can do it like this:
df[df['state'].apply(lambda state: state[0]=='W')].head()
```

```python
# The map method can be used to replace values in a column by 
# passing a dictionary of the form {old_value: new_value} as its argument:

d = {'no' : False, 'yes' : True}
df['international plan'] = df['international plan'].map(d)
df.head()
```

```python
# same thing with replace method
df = df.replace({'voice mail plan' : d})
df.head()
```

```python
# Grouping 
cols_to_show = ['total day minutes', 'total eve minutes']
df.groupby(['churn'])[cols_to_show].describe()
```

```python
# passing a list of functions to agg()
df.groupby(['churn'])[cols_to_show].agg([np.mean, np.std])
```

```python
# cross summary tables 
# to see how the observations in our sample are distributed 
# in the context of two variables 

pd.crosstab(df['churn'], df['international plan'])
```

```python
pd.crosstab(df['churn'], df['international plan'], normalize=True, )
```

```python
# pivot tables 
df.pivot_table(['total day calls', 'total night calls'], ['area code'], aggfunc='mean' )
```

```python
# dataframe transformations 
total_calls = df['total day calls'] + df['total eve calls'] + df['total night calls'] + df['total intl calls']
df.insert(loc=len(df.columns), column='total calls', value=total_calls)
```

```python
df.head()
```

```python
# other way of doing something similar
df['total charge'] = df['total day charge'] + df['total eve charge'] + df['total night charge'] + df['total intl charge']
df.head()
```

```python
# delete (drop) columns (axis = 1) or rows (axis = 0)
# inplace = True -> transform original matrix
df.drop(['total charge', 'total calls'], axis=1, inplace=True)
df.head()
```

```python
# deleting rows 
df.drop([1,2]).head()
```

```python
pd.crosstab(df['international plan'], df['churn'], margins=True)
```

```python
sns.countplot(x='international plan', data=df, hue='churn')
```

```python
pd.crosstab(df['churn'], df['customer service calls'], margins=True)
```

```python
sns.countplot(x='customer service calls', data=df, hue='churn')
```

```python
# adding a binary column for service calls > 3

df['many service calls'] = (df['customer service calls'] > 3).astype('int')
```

```python
pd.crosstab(df['many service calls'], df['churn'], margins=True)
```

```python
sns.countplot(x='many service calls', data=df, hue='churn')
```

```python
pd.crosstab(df['many service calls'] & df['international plan'], df['churn'])
```

Therefore, predicting that a customer is not loyal (Churn=1) in the case when the number of calls to the service center is greater than 3 and the International Plan is added (and predicting Churn=0 otherwise), we might expect an accuracy of 85.8% (we are mistaken only 464 + 9 times). This number, 85.8%, that we got with very simple reasoning serves as a good starting point (baseline) for the further machine learning models that we will build.
